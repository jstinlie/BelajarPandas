```python
#persiapan data frame
n_rows = 10
n_cols = 10
cols = tuple('ABCDEFGHIJ')

df = pd.DataFrame(np.random.randint(1, 20, size=(n_rows, n_cols)),
                 columns=cols)
df
```

```python
# mengganti nama untuk sebuah kolom pada data frame
df.rename(columns={'G' : 'Organisasi'})
```
