```python
# import pandas dan numpy
import pandas as pd
import numpy as np

print(np.__version__)
print(pd.__version__)
```

```python
# persiapan data frame
n_rows = 20
n_cols = 10
cols = tuple('ABCDEFGHIJ')

df = pd.DataFrame(np.random.randint(1, 40, size=(n_rows, n_cols)),
                  columns=cols)
df
```

```python
# seleksi kolom dan baris menggunakan loc
df.loc[[0,9,15], ['G' , 'I', 'J']]
```

```python
# seleksi baris dengan kondisi
df.loc[df['C']>20, ['C', 'G', 'I']]
```
