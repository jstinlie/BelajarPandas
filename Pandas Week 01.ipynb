```python
# pandas04
import pandas as pd
import numpy as np

n_rows = 10
n_cols = 2
cols = ['nilai_desimal', 'nilai_bulat']

df = pd.DataFrame(np.random.randint(70, 100, size=(n_rows, n_cols)),
                 columns=cols)
df['nilai_desimal'] = df['nilai_desimal'].astype('float')

df.index = pd.util.testing.makeDateIndex(n_rows, freq='H')
df = df.reset_index()

df['initial_name'] = list('ABCDEFGHIJ')

df
```

```python
df.select_dtypes(include='float') # memilih kolom bertipe data float
```

```python
df.select_dtypes(include='int') # memilih kolom bertipe data int
```

```python
df.select_dtypes(include='object') # memilih kolom bertipe data string atau objek
```

```python
df.select_dtypes(include='number') # memilih kolom bertipe data numerik
```

```python
df.select_dtypes(include=['float', 'object']) #memilih kombinasi tipe data float dan object
```

```python
df.select_dtypes(include=['int', 'object']) #memilih kombinasi tipe data int dan object
```
